# Supprimer les valeurs nulls

L'ensemble de données contient plusieurs valeurs nulles (non valides).il existe 3 facons de replacer les 0

• Ignore les

• Supprimer les lignes du dataframe qui contient 0

• Remplacez_les par d'autres valeurs(Imputation): utiliser l'imputation moyenne ou remplacer par la médiane

nous allons remplacer les zéros par la moyenne

```python
#replacer les zéros avec la moyenne(mean)
diabetes['Insulin']=diabetes['Insulin'].replace(0,diabetes['Insulin'].mean())
print('ya pas des valeurs nulls dans Insulin ',diabetes[diabetes['Insulin']==0].shape[0])
```

```python
diabetes['Glucose']=diabetes['Glucose'].replace(0,diabetes['Glucose'].mean())
diabetes['SkinThickness']=diabetes['SkinThickness'].replace(0,diabetes['SkinThickness'].mean())
diabetes['BloodPressure']=diabetes['BloodPressure'].replace(0,diabetes['BloodPressure'].mean())
diabetes['BMI']=diabetes['BMI'].replace(0,diabetes['Insulin'].mean())
```

# detecter et supprimer  les valeurs aberrantes

il y'a des valeurs un peu anormales comme la valeur maximale de l'insuline égale à 846.donc c'est une valeur aberrante et nous utilisons des boites à moustaches pour confirmer

```python
diabetes.boxplot(figsize=(15,5))
#diabetes[['Glucose' , 'BloodPressure' , 'SkinThickness' , 'Insulin' ,'BMI' ,'Age']].boxplot(figsize=(15,5))
```

les boîtes courtes signifient que leurs points de données planent constamment autour des valeurs du centre. Les boîtes plus hautes impliquent des données plus variables.

pour supprimer les valeurs aberrantes on utilise Score IQR(IQR est la difference entre le 75e et le 25e percentile qui est le quartiles supérieur et inférieur)

```python
Q1 = diabetes.quantile(0.25)
Q3 = diabetes.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

il s'agit du score IQR pour chaque colonne

```python
diabetes = diabetes[~((diabetes < (Q1 - 1.5 * IQR)) |(diabetes >(Q3 +1.5 * IQR))).any(axis=1)]
diabetes.shape
```

```python
from sklearn.metrics import confusion_matrix
```

```python
# Method to plot the confusion matrix
def plot_confusion_matrix(cm, classes, normalize=False, title='Confusion matrix', cmap=plt.cm.Blues):

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
```

```python
import tensorflow as tf
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
model = tf.keras.models.Sequential()
model.add(Dense(64, activation='relu',input_dim=8)),
model.add(Dense(64,activation='relu'))
model.add(Dense(14,activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
```
